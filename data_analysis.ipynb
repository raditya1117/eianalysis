```python
positives={"Always":5,
            "Most of the time":4,
           "Frequently":3,
            "Sometimes":2,
            "Rarely":1,
            "Never":0,
}
negatives={"Always":-5,
            "Most of the time":-4,
           "Frequently":-3,
            "Sometimes":-2,
            "Rarely":-1,
            "Never":0,
}
emotions=['fear', 'happiness', 'anger',
       'embarrassment', 'pride', 'surprise', 'contempt', 'disgust', 'flirt',
       'pain', 'compassion', 'amusement', 'interest', 'sadness', 'desire']
```

```python
def get_score(column_name,data):
    if column_name=="fear":
        data["fear"]=data["fear"].apply(lambda x: 1 if x=="Fear" else 0 )
    elif column_name=="happiness":
        data["happiness"]=data["happiness"].apply(lambda x: 1 if x=="Happiness" else 0 )
    elif column_name=="anger":
        data["anger"]=data["anger"].apply(lambda x: 1 if x=="Anger" else 0 )
    elif column_name=="embarrassment":
        data["embarrassment"]=data["embarrassment"].apply(lambda x: 1 if x=="Embarrassment" else 0 )
    elif column_name=="pride":
        data["pride"]=data["pride"].apply(lambda x: 1 if x=="Pride" else 0 )
    elif column_name=="surprise":
        data["surprise"]=data["surprise"].apply(lambda x: 1 if x=="Surprise" else 0 )
    elif column_name=="contempt":
        data["contempt"]=data["contempt"].apply(lambda x: 1 if x=="Contempt" else 0 )
    elif column_name=="disgust":
        data["disgust"]=data["disgust"].apply(lambda x: 1 if x=="Disgust" else 0 )
    elif column_name=="flirt":
        data["flirt"]=data["flirt"].apply(lambda x: 1 if x=="Flirtatiousness" else 0 )
    elif column_name=="pain":
        data["pain"]=data["pain"].apply(lambda x: 1 if x=="Pain" else 0 )
    elif column_name=="compassion":
        data["compassion"]=data["compassion"].apply(lambda x: 1 if x=="Compassion" else 0 )
    elif column_name=="amusement":
        data["amusement"]=data["amusement"].apply(lambda x: 1 if x=="Amusement" else 0 )
    elif column_name=="interest":
        data["interest"]=data["interest"].apply(lambda x: 1 if x=="Interest" else 0 )
    elif column_name=="sadness":
        data["sadness"]=data["sadness"].apply(lambda x: 1 if x=="Sadness" else 0 )
    elif column_name=="desire":
        data["desire"]=data["desire"].apply(lambda x: 1 if x=="Desire" else 0 )
    return data
```

```python
for column_name in emotions:
    data=get_score(column_name,data)
```
